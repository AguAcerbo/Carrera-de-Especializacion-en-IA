### Trabajo Practico 2
<p style='text-align: justify;'>
    En este trabajo practico buscaremos crear una compresión de una imagen de extensión jpg a través de la descomposición en valores singulares (SVD).
    El teorema de la descomposición de valores singulares establece que dada una matriz rectangular $A^{mxn}$ de rango $r \in [0, min(m,n)]$, se puede descomponerse de la siguiente forma:
</p>

<p style='text-align: center;'>
    $A^{mxn} = U^{mxm} . \Sigma^{mxn} . V^{T nxn}$
</p>

<p style='text-align: justify;'>
    Donde $U \in R ^{mxm}$ es una matriz ortogonal con vectores columnas $u_i, i = 1, ..., m$ y $V \in R ^{nxn}$ es una matriz ortogonal con vectores columnas $v_j, j = 1, ..., n$. Por otro lado, $\Sigma$ es una matriz m x n con $\Sigma_{ii} = \sigma_i \geqslant 0$ y $\Sigma_{ij} = 0$ para todo $i \neq j$. Los valores $\sigma_i$ se denominan valores singulares y estan ordinados de manera descendentes, es decir, $\sigma_1 \geqslant \sigma_2 \geqslant ... \geqslant \sigma_r \geqslant 0$.
    
Los valores singulares de la matriz $\Sigma$ son unicos pero para que esta matriz tenga las mismas dimensiones que $A$ debe ser extendida con valores ceros según sea necesario, por lo que tendremos matrices de la forma:
</p>

<p style='text-align: center;'>
    $\Sigma = \begin{bmatrix} \sigma_1 & 0 & 0 \\ 0 & ^.._. & 0 \\ 0 & 0 & \sigma_n \\ 0 & ... & 0 \\ : & ^.._. & : \\ 0 & ... & 0 \end{bmatrix}$ Si $m > n$
</p>

O bien:

<p style='text-align: center;'>
    $\Sigma = \begin{bmatrix} \sigma_1 & 0 & 0 & 0 & ... & 0\\ 0 & ^.._. & 0 & 0 & ^.._. & 0 \\ 0 & 0 & \sigma_m & 0 & ... & 0\end{bmatrix}$     Si $m < n$
</p>

<p style='text-align: justify;'>
    A continuación utilizaremos lo antes expuesto para representar la matriz $A$ como una suma de matrices de menor rango, lo que nos permitirá obtener una aproximación de $A$ que nos ahorrará recursos computacionales.
    </p>

```python
from matplotlib.image import imread
from matplotlib.image import imsave
import matplotlib.pyplot as plt
import numpy as np
```

<p style='text-align: justify;'>
    En primer lugar, a partir de la lectura de la imagen, obtenemos $A$, y mostraremos a reducir.
</p>

```python
img_orig = imread('Uvas.jpg')
A = img_orig[:,:,0]
img = plt.imshow(A)
img.set_cmap('gray')

plt.show()

print(f"Dimensiones de A: {A.shape}")
```

<p style='text-align: justify;'>
    A partir de lo obtenido vemos que la matriz A posee dimensiones 2268x4032 (por lo tanto el rango es 2268). El número de columnas es mayor al numero de filas $(m < n)$, por lo que podemos plantear:
</p>

<p style='text-align: center;'>
    $A =
    \begin{bmatrix} u_1 & u_2 & ... & u_m \end{bmatrix}
    \begin{bmatrix} \sigma_1 & 0 & 0 \\ 0 & ^.._. & 0 \\ 0 & 0 & \sigma_m \\ 0 & ... & 0 \\ : &  & : \\ 0 & ... & 0 \end{bmatrix}
    \begin{bmatrix} v^*_1 \\ v^*_2 \\ : \\ v^*_n \end{bmatrix} 
    =
    \sum \limits _{k=1} ^{m} \sigma_k u_k v^*_k$
</p>

<p style='text-align: justify;'>
    De una manera más general se puede escribir utilizando el rango de $A$, donde $r = rank(A) = m$, como:
</p>

<p style='text-align: center;'>
    $A = \Large \sum \limits _{k=1} ^{r} \sigma_k u_k v^*_k$
</p>

<p style='text-align: justify;'>
    Para obtener las matrices de descomposición utilizaremos la función svd que pertenece al conjunto de funciones de algebra lineal de la libreria numpy. Esta funcion nos devolverá 3 matrices, la matriz U, una matriz S de dimensiones (rx1), siendo r el rango de nuestra matriz A, con los r valores singulares de la matriz a descomponer y por último devolverá la matriz VT.
    
A partir de la matriz S, formaremos la matriz diagonal $\Sigma$ para los cálculos posteriores.
</p>


```python
U, S, VT = np.linalg.svd(A)
Sigma = np.diag(S)

print(f"Dimensiones de U: {U.shape}")
print(f"Dimensiones de S: {S.shape}")
print(f"Dimensiones de Sigma: {Sigma.shape}")
print(f"DImensiones de VT: {VT.shape}")
```

<p style='text-align: justify;'>
    Podemos corroborar las dimensiones de las matrices obtenidas, con la excepción de la matriz $\Sigma$, ya que esta es la matriz diagonal de los valores singulares sin la extención de ceros antes mencionada.
    
A partir de estas matrices ahora podemos comenzar a probar la reducción de la matriz $A$. Habiamos visto que podiamos escribir:
</p>

<p style='text-align: center;'>
    $A = \Large \sum \limits _{i=1} ^{r} \sigma_i u_i v^*_i$
</p>

<p style='text-align: justify;'>
    De esta expresión modificaremos el el limite de la sumatoria reduciendolo. Por lo tanto, definiremos la "aproximación de rango-$k$ de la matriz $A$" como:
</p>

<p style='text-align: center;'>
    $\hat A(k) = \Large \sum \limits _{i=1} ^{k} \sigma_i u_i v^*_i$, tal que $k < r$
</p>

<p style='text-align: justify;'>
    A continuación calcularemos las matrices de aproximación para varios valores de $k$, mostrando la imagen obtenida a partir de las aproximaciones realizadas:
</p>

```python
def Aprox_SVD(U, Sigma, VT, k):
    """
    Esta función calcula la aproximación de rango k de una matriz rectangular
    haciendo uso de la descomposición de valores singulares
    """
    Ared = U[:,:k] @ Sigma[:k,:k] @ VT[:k,:]
    img1 = plt.imshow(Ared)
    img1.set_cmap('gray')
    plt.show()
    return Ared
```

#### Aproximación de A, rango-1:

```python
k = 1
Ar1 = Aprox_SVD(U, Sigma, VT, k)
imsave('Uvas_rank1.jpg',Ar1,cmap="gray")
```

#### Aproximación de A, rango-5:

```python
k5 = 5
Ar5 = Aprox_SVD(U, Sigma, VT, k5)
imsave('Uvas_rank5.jpg',Ar5,cmap="gray")
```

#### Aproximación de A, rango-25:

```python
k = 25
Ar25 = Aprox_SVD(U, Sigma, VT, k)
imsave('Uvas_rank25.jpg',Ar25,cmap="gray")
```

#### Aproximación de A, rango-50:

```python
k = 50
Ar50 = Aprox_SVD(U, Sigma, VT, k)
imsave('Uvas_rank50.jpg',Ar50,cmap="gray")
```

#### Aproximación de A, rango-100:

```python
k = 100
Ar100 = Aprox_SVD(U, Sigma, VT, k)
imsave('Uvas_rank100.jpg',Ar100,cmap="gray")
```

#### Aproximación de A, rango-500:

```python
k = 500
Ar500 = Aprox_SVD(U, Sigma, VT, k)
imsave('Uvas_rank500.jpg',Ar500,cmap="gray")
```
