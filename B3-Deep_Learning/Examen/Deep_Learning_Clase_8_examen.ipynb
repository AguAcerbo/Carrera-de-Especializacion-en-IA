## Ejercicio 1



Se quiere encontrar el máximo de la siguiente función:

$z = -(x - 2)^2 - (y - 3)^2 + 4$

(a) Aplicar gradiente e igualar a zero para encontrar los valores de x e y donde z tiene un máximo. Cuál es el valor del máximo?

(b) Aplicar SGD para encontrar la ubicación del máximo de manera numérica.

$dz/dy = -2(y - 3) -> dz/dy = 0 si y = 3$

$dz/dx = -2(y - 2) -> dz/dx = 0 si x = 2$

Por lo tanto el punto maximo de la funcion se encuentra en la coordenada $x,y =(2,3)$

Donde z toma un valor de $Z(2,3) = 4$

```python
x = np.arange(-5, 10, 0.01)
y = np.arange(-5, 10, 0.01)
z = (-1)*(x-2)**2-(y-3)**2+4
```

```python
def max_z(lr=0.01, amt_epochs=100):
    """
    z = (-1)*(x-2)**2-(y-3)**2+4
    """
    x = []
    y = []
    epch = []
    loss = []
    x_min = np.random.randn()
    y_min = np.random.randn()
    for epoch in range(amt_epochs):
        x_gradiente = -2*(x_min-2)          #derivada de z respecto de x
        x_min = x_min + (lr * x_gradiente)
        x.append(x_min)

        y_gradiente = -2*(y_min-3)          #derivada de z respecto de y
        y_min = y_min +(lr * y_gradiente)
        y.append(y_min)

        epch.append(epoch)
        loss.append(x_gradiente+y_gradiente)
        
        if (abs(x_gradiente+y_gradiente) < 0.0001):
            print(f"Se detiene calculo por bajo error: {x_gradiente+y_gradiente}")
            break
            
        if (abs(lr * x_gradiente+y_gradiente) < 0.00001):
            print(f"Se detiene calculo por baja actualizacion del valor buscado {abs(lr * x_gradiente+y_gradiente)}")
            break
    return epch, x, y, loss
```

```python
def calc_null(dta):
    """
    Esta función calcula la cantidad numerica y porcentual de
    datos faltantes de cada columna del dataset de entrada.
    """
    cant_null = []
    porc_null = []
    for feature in dta.columns:
        cant_null.append(sum(dta[feature].isnull()))
        porc_null.append(sum(dta[feature].isnull())*100/dta.shape[0])
    return cant_null, porc_null 

cant_null, porc_null = calc_null(df_train)
orden =  np.argsort(porc_null)[::-1]

print("Datos Faltantes:")
for i in range(len(orden)):
    if cant_null[orden[i]] != 0:
        print(f"{df_train.columns[orden[i]]}:\tDatos Nan: {cant_null[orden[i]]}\tEn%: {porc_null[orden[i]]}")
```

### (g) Deep learning sin embeddings

1. Entrenar un modelo de Deep Learning que no utilice embeddings (no usar User_Id y Product_Id).

```python
device = ""
if torch.cuda.is_available():
    device = "cuda:0"
else:
    device = "cpu"
    
print(torch.cuda.is_available())
```

```python
class CustomDataset(Dataset):
    def __init__(self, X, Y):
        super().__init__()
        self.X = X
        self.Y = Y

    def __len__(self):
        return self.X.shape[0]

    def __getitem__(self, idx):
        return self.X[idx,:], self.Y[idx]
```

```python
training = CustomDataset(x_train, y_train)
validation = CustomDataset(x_valid, y_valid)
testing = CustomDataset(x_test, y_test)
```

```python
print(len(training))
print(len(validation))
print(len(testing))
```

```python
training_dataloader = DataLoader(training, batch_size=128, shuffle=True)
validation_dataloader = DataLoader(validation,batch_size= 128, shuffle=True)
test_dataloader = DataLoader(testing, batch_size=128, shuffle=True)
```

```python
class NNet(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.linear_1 = torch.nn.Linear(in_features=93, out_features=200, bias=True)
        self.relu_1 = torch.nn.ReLU()
        self.linear_2 = torch.nn.Linear(in_features=200, out_features=100, bias=True)
        self.relu_2 = torch.nn.ReLU()
        self.linear_3 = torch.nn.Linear(in_features=100, out_features=1, bias=True)
    
    def forward(self, x):
        z1 = self.linear_1(x)
        a1 = self.relu_1(z1)
        z2 = self.linear_2(a1)
        a2 = self.relu_2(z2)
        y = self.linear_3(a2)
        return y
```

```python
nnet = NNet()
```

```python
print(nnet)
```

```python
criterion = torch.nn.BCEWithLogitsLoss(reduction='sum')
optimizer = torch.optim.Adam(nnet.parameters(), lr=0.01)
```

```python
nnet.to(device)

for epoch in range(100):

    running_loss = 0
    nnet.train()
    for i, data in enumerate(training_dataloader):

        # batch
        x, y = data
        x = x.to(device).float()
        y = y.to(device).float().reshape(-1,1)

        # set gradient to zero
        optimizer.zero_grad()

        # forward 
        y_hat = nnet(x)

        # loss
        loss = criterion(y_hat, y)

        # backward
        loss.backward()

        # update of parameters
        optimizer.step()

        # compute metrics and statistics
        running_loss += loss.item()
    
    nnet.eval()
    with torch.no_grad():
        nnet_test_scores = []
        truth = []
        for i, data in enumerate(test_dataloader):
            # batch
            x, y = data
            x = x.to(device).float()
            y = y.to(device).float().reshape(-1,1)

            # forward 
            y_hat = nnet(x)
            y_hat = torch.sigmoid(y_hat)

            # accumulate data
            truth += list(y.cpu().detach().numpy()) 
            nnet_test_scores += list(y_hat.cpu().detach().numpy())

        fpr, tpr, thresholds = metrics.roc_curve(truth, nnet_test_scores)
        auc = metrics.auc(fpr, tpr)
        print(f"Epoch = {epoch} | loss = {running_loss / len(training)} | auc = {auc}")
```

### (h) Deep learning con embeddings

1. Entrenar un model de Deep Learning que utilice embeddings de los usuarios y los productos.

```python
class CustomDatasetWithEmbedding(Dataset):
    def __init__(self, X, user_idx, product_idx, Y):
        super().__init__()
        self.user_idx = user_idx
        self.product_idx = product_idx
        self.X = X
        self.Y = Y

    def __len__(self):
        return self.X.shape[0]

    def __getitem__(self, idx):
        return self.X[idx,:], self.user_idx[idx], self.product_idx[idx], self.Y[idx]
```

```python
training_emb = CustomDatasetWithEmbedding(x_train, user_index_train, product_index_train, y_train)
validation_emb = CustomDatasetWithEmbedding(x_valid, user_index_valid, product_index_valid, y_valid)
testing_emb = CustomDatasetWithEmbedding(x_test, user_index_test, product_index_test, y_test)
```

```python
training_dataloader_emb = DataLoader(training_emb, batch_size=64, shuffle=True)
validation_dataloader_emb = DataLoader(validation_emb, batch_size=64, shuffle=True)
test_dataloader_emb = DataLoader(testing_emb, batch_size=32, shuffle=True)
```

```python
class NNet_emb(torch.nn.Module):
    def __init__(self, number_of_users, number_of_products, embedding_dim1, embedding_dim2):
        super().__init__()
        self.embedding_users = torch.nn.Embedding(num_embeddings=number_of_users, embedding_dim=embedding_dim1)
        self.embedding_products = torch.nn.Embedding(num_embeddings=number_of_products, embedding_dim=embedding_dim2)
        self.linear_1 = torch.nn.Linear(in_features=(93 + embedding_dim1 + embedding_dim2), out_features=200, bias=True)
        self.relu_1 = torch.nn.ReLU()
        self.linear_2 = torch.nn.Linear(in_features=200, out_features=300, bias=True)
        self.relu_2 = torch.nn.ReLU()
        self.linear_3 = torch.nn.Linear(in_features=300, out_features=200, bias=True)
        self.relu_3 = torch.nn.ReLU()
        self.linear_4 = torch.nn.Linear(in_features=200, out_features=1, bias=True)

    def forward(self, x, user_idx, product_idx):
        user_emb = self.embedding_users(user_idx)
        product_emb = self.embedding_products(product_idx)
        final_input = torch.cat([x, user_emb, product_emb], dim=1)
        z1 = self.linear_1(final_input)
        a1 = self.relu_1(z1)
        z2 = self.linear_2(a1)
        a2 = self.relu_2(z2)
        z3 = self.linear_3(a2)
        a3 = self.relu_3(z3)
        y = self.linear_4(a3)
        return y
```

```python
nnet_emb = NNet_emb(number_of_users=len(unique_users),number_of_products=len(unique_products), embedding_dim1=32, embedding_dim2=16)
```

```python
print(nnet_emb)
```

```python
criterion = torch.nn.BCEWithLogitsLoss(reduction='sum')
optimizer = torch.optim.Adam(nnet_emb.parameters(), lr=0.01)
```

```python
nnet_emb.to(device)

for epoch in range(100):

    running_loss = 0
    nnet_emb.train()
    for i, data in enumerate(training_dataloader_emb):

        # batch
        x, user_idx, product_idx, y = data

        x = x.to(device).float()
        y = y.to(device).float().reshape(-1,1)

        # set gradient to zero
        optimizer.zero_grad()

        # forward
        y_hat = nnet_emb(x, user_idx.to(device), product_idx.to(device))

        # loss
        loss = criterion(y_hat, y)

        # backward
        loss.backward()

        # update of parameters
        optimizer.step()

        # compute metrics and statistics
        running_loss += loss.item()
    
    nnet_emb.eval()
    with torch.no_grad():
        nnet_test_scores = []
        truth = []
        for i, data in enumerate(test_dataloader_emb):
            # batch
            x, user_idx, product_idx, y = data
            x = x.to(device).float()
            y = y.to(device).float().reshape(-1,1)

            # forward 
            y_hat = nnet_emb(x, user_idx.to(device), product_idx.to(device))
            y_hat = torch.sigmoid(y_hat)

            # accumulate data
            truth += list(y.cpu().detach().numpy()) 
            nnet_test_scores += list(y_hat.cpu().detach().numpy())

        fpr, tpr, thresholds = metrics.roc_curve(truth, nnet_test_scores)
        auc = metrics.auc(fpr, tpr)
        print(f"Epoch = {epoch} | loss = {running_loss / len(training_emb)} | auc = {auc}")
```
