# Muestreo aleatorio con SciPy

Plan:
- Muestreos aleatorios con distintas distribuciones utilizando SciPy.
- Diseño de experimentos. Ejemplos.

```python
#import matplotlib.pyplot as plt
# Usaremos Seaborn para visualizar las distribuciones (también podemos hacerlo con Matplotlib)
import seaborn as sns; sns.set(rc={'figure.figsize':(22,4)})
```

```python
N_SAMPLES = 10000
```

### Distribución uniforme

```python
from scipy.stats import uniform

start = 10
width = 20

# RVS = Random Variable Sampling
data_uniform = uniform.rvs(size=N_SAMPLES, loc = start, scale=width)
```

```python
type(data_uniform), data_uniform.ndim, data_uniform.shape
```

```python
ax = sns.distplot(data_uniform,
                  bins=100,
                  kde=True )
ax.set(xlabel='Distribución Uniforme', ylabel='Frecuencia')
```

### Distribución normal

```python
from scipy.stats import norm

mean = 0
std_dev = 1

data_norm = norm.rvs(size=N_SAMPLES,loc=mean,scale=std_dev)
```

```python
ax = sns.distplot(data_norm,
                  bins=100,
                  kde=True )
ax.set(xlabel='Distribución Normal', ylabel='Frecuencia')
```

### Distribución gamma

```python
from scipy.stats import gamma

a = 5
data_gamma = gamma.rvs(a=5, size=N_SAMPLES)
```

```python
ax = sns.distplot(data_gamma,
                  bins=100,
                  kde=True )
ax.set(xlabel='Distribución Gamma', ylabel='Frecuencia')
```

### Distribución exponencial

```python
from scipy.stats import expon

lambda_param = 1

data_expon = expon.rvs(scale=1/lambda_param,loc=0,size=N_SAMPLES)
```

```python
ax = sns.distplot(data_expon,
                  bins=100,
                  kde=True )
ax.set(xlabel='Distribución Exponencial', ylabel='Frecuencia')
```

### Distribución de Poisson

```python
from scipy.stats import poisson

mean= 3

data_poisson = poisson.rvs(mu=3, size=N_SAMPLES)
```

```python
ax = sns.distplot(data_poisson,
                  bins=30,
                  kde=False )
ax.set(xlabel='Distribución de Poisson', ylabel='Frecuencia')
```

### Distribución binomial

```python
from scipy.stats import binom

data_binom = binom.rvs(n=10,p=0.8,size=N_SAMPLES)
```

```python
ax = sns.distplot( data_binom, kde=False )
ax.set(xlabel='Distribución Binomial', ylabel='Frecuencia')
```

### Distribución de Bernoulli

```python
from scipy.stats import bernoulli

data_bernoulli = bernoulli.rvs(size=N_SAMPLES,p=0.6)
```

```python
ax= sns.distplot(data_bernoulli, kde=False)
ax.set(xlabel='Bernoulli Distribution', ylabel='Frequency')
```
