```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mu = 0
sigma = 1
N = 100000
degFreedom = 5
bins = 50
```

```python
# Random values data series
X = np.random.normal(mu, sigma, N)

Y = np.random.standard_t(degFreedom, N)
```

```python
# Plot
plt.plot(X)
plt.show()

count, bins, ignored =plt.hist(X, bins, density=True)
plt.plot(bins, 1/(sigma*np.sqrt(2*np.pi))*np.exp(-(bins-mu)**2/(2*sigma**2)), linewidth = 2, color='r')
plt.show()
```

```python
# Data series
fig, (ax1, ax2) = plt.subplots(2)
ax1.set_title('X vs. Y timeseries')
ax1.plot(X)
ax2.plot(Y)
plt.show()
```

```python
# Histograms
fig, (ax1, ax2) = plt.subplots(2)
ax1.set_title('X vs. Y histograms')
ax1.hist(X, bins=30)
ax2.hist(Y, bins=30)
plt.show()
```

```python
# using Dataframe structure
df=pd.DataFrame({'Y':Y, 'X':X})
df.plot()
plt.show()
```

```python
## Stats
df.X.describe
df.Y.describe()
dt = pd.DataFrame(df.X.describe())
dt = pd.concat([df.X.describe(), df.Y.describe()],axis=1)
```
