# Anotaciones en imágenes 

Funciones básicas para graficar líneas, círculos y texto

```python
# OpenCV-Python utiliza NumPy para el manejo de imágenes
import numpy as np
# cv2 es el módulo python para acceder a OpenCV 
import cv2 as cv
# Usamos las poderosas herramientas de graficación de matplotlib para mostrar imágenes, perfiles, histogramas, etc
import matplotlib.pyplot as plt

#Si queremos que las imágenes sean mostradas en una ventana emergente quitar el inline
%matplotlib inline
```

```python
img = cv.imread('imgBloque1.bmp')
img_rectangle = img.copy()
img_ellipse = img.copy()
img_circle = img.copy()
img_polylines = img.copy()

# Dibujo una línea
# line(imagen,inicio,fin,color,ancho_linea,tipo_linea)
cv.line(img,(350,0),(350,410),(0,0,0),2,cv.LINE_AA)
cv.line(img,(250,0),(350,410),(0,0,255),2,cv.LINE_AA)
plt.imshow(img[...,::-1])
```

#### Tipos de líneas
Filled: cv.FILLED
4-connected line: cv.LINE_4 
8-connected line: cv.LINE_8
Antialiased line: cv.LINE_AA

#### Ejercicio:
Dibujar rectángulos (cv.rectangle)
elipses (cv.ellipse)
círculos (cv.circle)
polígonos (cv.polylines)

```python
# Dibujo un rectangulo
# rectangle(imagen,inicio,fin,color,ancho_linea,tipo_linea)
cv.rectangle(img_rectangle,(350,100),(200,410),(0,255,0),2,cv.LINE_AA)
cv.rectangle(img_rectangle,(250,50),(500,450),(255,0,255),10,cv.LINE_AA)
plt.imshow(img_rectangle[...,::-1])
```

```python
# Dibujo un ellipse
# ellipse(imagen, center_coordinates, axesLength, angle, startAngle, endAngle, color, ancho_linea, tipo_linea)

center_coordinates = (150, 100)
axesLength = (100, 50)
angle = 0
startAngle = 90
endAngle = 360
cv.ellipse(img_ellipse, center_coordinates, axesLength, angle, startAngle, endAngle,(0,0,255),2)

cv.ellipse(img_ellipse,(150,300),(100,100),0,0,360,(255,0,0),2,cv.LINE_AA)

cv.ellipse(img_ellipse,(500,250),(100,200),45,0,360,(255,255,0),5,cv.LINE_AA)

plt.imshow(img_ellipse[...,::-1])
```

```python
# Dibujo un circle
# circle(imagen,center,radius,color,ancho_linea,tipo_linea)

cv.circle(img_circle, (250,300), 150, (255,0,0), 2, cv.LINE_AA)

cv.circle(img_circle, (250,300), 50, (0,255,0), 5, cv.LINE_AA)

cv.circle(img_circle, (150,100), 10, (0,0,255), 10, cv.LINE_AA)

plt.imshow(img_circle[...,::-1])
```

```python
# Dibujo un polylines
# polylines(imagen,center,radius,color,ancho_linea,tipo_linea)

pts = np.array([[250, 70], [250, 160], 
                [110, 200], [200, 160], 
                [200, 70], [110, 20]],
               np.int32)
  
pts = pts.reshape((-1, 1, 2))

cv.polylines(img_polylines, [pts], True, (255,0,255), 2, cv.LINE_AA)

#cv.circle(img_polylines, (250,300), 50, (0,255,0), 5, cv.LINE_AA)

#cv.circle(img_polylines, (150,100), 10, (0,0,255), 10, cv.LINE_AA)

plt.imshow(img_polylines[...,::-1])
```
